# Fix INMET files format

```python
import shutil
from unidecode import unidecode
from pathlib import Path
import os


def get_ascii_char(c: str) -> str:
        decoded = unidecode(c)
        if len(decoded) > 1:
            return ''
        ascii_number = ord(decoded)
        ascii_char = bytes([ascii_number]).decode("ascii")
        return ascii_char

def get_ascii_content(content: str) -> str:
    ascii_content = ''
    for c in content:
        ascii_char = get_ascii_char(c)
        ascii_content += ascii_char

    return ascii_content


for root, _, files in os.walk(paths.raw_inmet_folder):
    for file in files:
        if not file.lower().endswith('.csv'):
            continue

        file_path = os.path.join(root, file)
        new_file_path = file_path.replace(paths.raw_path, paths.raw_fixed_path)
        Path(new_file_path).parent.mkdir(parents=True, exist_ok=True)

        with open(file_path, 'r', encoding='latin-1') as latin:
            content = latin.read()
            with open(new_file_path, 'w', encoding='utf-8') as fixed:
                fixed.write(get_ascii_content(content))
```

# Remove special chars INPE files

```python
import unicodedata
import os


for root, _, files in os.walk(paths.raw_inpe_folder):
    for file in files:
        if not file.lower().endswith('.csv'):
            continue

        file_path = os.path.join(root, file)
        with open(file_path, 'r', encoding='utf-8') as f:
            content = f.read()
        content = unicodedata.normalize("NFKD", content).encode("ascii", "ignore").decode("ascii")
        new_file_path = file_path.replace(paths.raw_path, paths.raw_fixed_path)
        with open(new_file_path, 'w', encoding='utf-8') as f:
            f.write(content)
```
